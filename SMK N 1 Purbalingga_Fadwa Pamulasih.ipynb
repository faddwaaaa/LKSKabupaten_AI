```python
import pathlib
import numpy as np

dir_path = pathlib.Path("Jenis Daun/Daun/training")
class_names = np.array(sorted([item.name for item in dir_path.glob("*")]))
print(class_names)
```

```python
import random
import matplotlib.pyplot as plt
import matplotlib.image as pltimg

def view_random_image(target_dir, target_class):
  target_folder = os.path.join(target_dir, target_class)
  random_image = random.choice(os.listdir(target_folder))

  img_path = os.path.join(target_folder, random_image)
  img = pltimg.imread(img_path)

  plt.imshow(img)
  plt.title(target_class)
  plt.axis(False)

  print(f"Image Shape : {img.shape}")
  return img
```

di bawah ini merupakan kode yang digunakan untuk menampilkan 1 gambar dari masing masing folder secara random(acak)

```python
img_dict = {}

for target_class in class_names:
  img = view_random_image('Jenis Daun/Daun/training', target_class)
  img_dict[target_class] = img

  plt.show()
```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen_augmented = ImageDataGenerator(rescale = 1/255.,
                                             rotation_range = 15,
                                             width_shift_range = 0.1,
                                             height_shift_range = 0.1,
                                             shear_range = 0.1,
                                             zoom_range = 0.1,
                                             horizontal_flip = True,
                                             fill_mode = 'nearest')
train_datagen = ImageDataGenerator(rescale = 1/255.)
valid_datagen = ImageDataGenerator(rescale = 1/255.)
```

```python
train_data_augmented = train_datagen_augmented.flow_from_directory('Jenis Daun/Daun/training',
                                                                   batch_size = 32,
                                                                   target_size = (224,224),
                                                                   class_mode = 'categorical',
                                                                   seed = 42,
                                                                   shuffle = True,
                                                                   subset = 'training')
train_data = train_datagen.flow_from_directory('Jenis Daun/Daun/training',
                                            batch_size = 32,
                                            target_size = (224,224),
                                            class_mode = 'categorical',
                                            seed = 42,
                                            shuffle = True,
                                            subset = 'training')
valid_data = valid_datagen.flow_from_directory('Jenis Daun/Daun/validation',
                                               batch_size = 32,
                                               target_size = (224,224),
                                               class_mode = 'categorical',
                                               seed = 42,
                                               shuffle = True,
                                               subset = 'training')
```

```python
images, labels = next(train_data)
augmented_images, augmented_labels = next(train_data_augmented)

print(f"{len(images), len(labels)}")
print(f"{len(augmented_images), len(augmented_labels)}")
```

```python
train_data_augmented_shuffled = train_datagen_augmented.flow_from_directory('Jenis Daun/Daun/training',
                                                                   batch_size = 32,
                                                                   target_size = (224,224),
                                                                   class_mode = 'categorical',
                                                                   seed = 42,
                                                                   shuffle = True,
                                                                   subset = 'training')
```

mengimport library tensorflow yang akan digunakan pada kode selanjutnya

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Conv2D, MaxPool2D, Dropout, Flatten, Dense, GlobalMaxPool2D
from tensorflow.keras.callbacks import ModelCheckpoint, EarlyStopping, ReduceLROnPlateau
```

Disini kita membuat variabel variabel untuk nantinya bisa dipakai pada pembuatan model (optional atau boleh dipakai boleh tidak), contohnya pada model 1 tidak menggunakan semua yang ada dibawah, tetapi nanti pada pembuatan model 2 menggunakan checkpoint dan early_stop

```python
checkpoint = ModelCheckpoint('best.h5', verbose = 1, save_best_only = True)
early_stop = EarlyStopping(patience = 4)
reduce_lr = ReduceLROnPlateau(monitor = 'val_loss', factor = 0.2, patience = 5, min_lr = 0.001)
```

pembuatan model dengan 6 layer dan 30 epoch

```python
model = Sequential([
    Conv2D(64, 3, activation = 'relu', input_shape = (224, 224, 3)),
    MaxPool2D(2),
    Conv2D(128, 3, activation = 'relu'),
    MaxPool2D(2),
    Flatten(),
    Dense(3, activation = 'softmax')
])

model.compile(
    loss = 'categorical_crossentropy',
    optimizer = 'adam',
    metrics = ['accuracy']
)

history_1 = model.fit(train_data_augmented_shuffled,
                      epochs = 30,
                      steps_per_epoch = len(train_data_augmented_shuffled),
                      validation_data = valid_data,
                      validation_steps = len(valid_data),
                      verbose = True)
```

kode dibawah berikut untuk menampilkan grafik model yang telah dibuat

```python
def plot_loss_curves(history):
  loss = history.history['loss']
  val_loss = history.history['val_loss']

  accuracy = history.history['accuracy']
  val_accuracy = history.history['val_accuracy']

  epochs = range(len(history.history['loss']))

  plt.plot( loss, label = 'training_loss')
  plt.plot( val_loss, label = 'validation_loss')
  plt.xlabel('Epoch')
  plt.title('Loss')
  plt.axis(False)
  plt.show()

  plt.figure()
  plt.plot( accuracy, label = 'training_accuracy')
  plt.plot( val_accuracy, label = 'validation_accuracy')
  plt.xlabel('Epoch')
  plt.title('Accuracy')
  plt.axis(False)
  plt.show()
```

Membuat gradio untuk menguji modelnya belajar atau tidak, misal kita mengupload foto daun jagung apakah ini daun jagung atau malah yang lainnya

```python
def classify_image(inp) :
  inp = tf.image.resize(inp, size = [224, 224])
  inp = tf.expand_dims(inp, axis = 0)
  inp = inp/255.

  prediction = model.predict(inp)
  prediction = np.round(prediction.flatten(), decimals = 2)
  print(prediction)

  confidences = {class_names[i]: (1-float(prediction) if i==0 else float(prediction)) for i in range (len(class_names))}
  return confidences

# Categorical Classification
gr.Interface(theme = gr.themes.Soft(),
             title = 'Klasifikasi Daun Jagung, Daun Padi dan Daun Tebu',
             description = '<center>Klasifikasi Daun Jagung, Daun Padi dan Daun Tebu Menggunakan Tensorflow dan Gradio</center>',
             article = 'Input Gambar Menggunakan Webcam, Upload atau Salin Gambar lalu Klik Submit untuk Menampilkan Hasilnya',
             fn = classify_image,
             inputs = gr.Image(inputs = 'image', label = 'Input Gambar'),
             outputs = gr.Label(num_top_classes = num_classes)).launch(share=True, debug=True)
```
